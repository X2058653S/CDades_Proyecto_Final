## Creación de nuevas variables

A partir de los datos disponibles en el dataset, nos parece interesante extraer y/o añadir la siguiente información :
- distrito : la variable 'location' representa los barrios de la ciudad, nos interesa crear la variable 'distrito' que incorpora los 10 distritos de la ciudad, ya que muy probablemente la variable 'location' con 74 ocurrencias no sea de aplicación para nuestros modelos de machine learning.
- cambio de precio : es evidente que el hecho de que una vivienda haya bajado de precio tiene un impacto en su precio final, nos planteamos extraer esta información de la variable 'precio'.
- nº de baños : esta información no existe como variable numérica, nos planteamos crearla a partir de la variable 'features'.

### Variable distrito

```python
df_Comprar['location'].nunique()
```

```python
distritos=pd.read_excel('distritos.xlsx')
distritos
```

```python
distritos.distrito.value_counts()
```

```python
df_Comprar=pd.merge(df_Comprar,distritos[['distrito','location']],on='location',how='left')
```

```python
df_Comprar.info()
```

```python
df_Comprar.distrito.value_counts()
```

### Variable cambio_precio

Este concepto se encuentra en la columna 'price' :

```python
df_Comprar['cambio_precio']= np.where(df_Comprar['price'].str.contains(r'ha bajado')==True,'Si','No')
```

```python
df_Comprar.cambio_precio.value_counts()
```

### Variable Baños

El número de baños aparece en la columna 'features' :

```python
df_Comprar['baños_pre']= np.where(df_Comprar['features'].str.contains(r'habitacion')==True,
                         df_Comprar['features'].str.split('-').str[2],
                        (np.where(df_Comprar['features'].str.contains(r'baño')==True,
                         df_Comprar['features'].str.split('-').str[1],'No')))
```

```python
df_Comprar['baños_pre'].value_counts()
```

Asignamos el número de baños que corresponde, en el caso de que no haya valor, asignamos el valor 1 por defecto al considerar que una vivienda tiene como mínimo un baño :

```python
df_Comprar['baños']= np.where(df_Comprar['baños_pre'].str.contains(r'baño')==True,
                          df_Comprar['baños_pre'].str.split(' ').str[1],1)
```

```python
df_Comprar.baños.value_counts()
```

Convertimos la variable a numérica :

```python
df_Comprar['baños'] = df_Comprar['baños'].astype('int')
```

```python
df_Comprar.drop(['baños_pre'], axis=1, inplace=True)
```

```python
df_Comprar.drop(['Unnamed: 0'], axis=1, inplace=True)
```
