## Creación de nuevas variables

A partir de los datos disponibles en el dataset, nos parece interesante extraer y/o añadir la siguiente información :
- distrito : la variable 'location' representa los barrios de la ciudad, nos interesa crear la variable 'distrito' que incorpora los 10 distritos de la ciudad, ya que muy probablemente la variable 'location' con 74 ocurrencias no sea de aplicación para nuestros modelos de machine learning.
- temporalidad : la modalidad de alquiler temporal, inferior a 11 meses, con toda seguridad tiene una influencia en el precio, nos planteamos extraer esta información de la variable 'description' del anuncio.
- cambio de precio : es evidente que el hecho de que una vivienda haya bajado de precio tiene un impacto en su precio final, nos planteamos extraer esta información de la variable 'precio'.
- nº de baños : esta información no existe como variable numérica, nos planteamos crearla a partir de la variable 'features'.

### Variable distrito

```python
df_Comprar['location'].nunique()
```

```python
barris=pd.DataFrame(df_Comprar['location'].value_counts())
barris
```

```python
barris.to_excel('barris.xlsx')
```

Recuperamos en excel el fichero barris.xlsx, creamos una columna con el nombre del distrito de cada barrio y creamos el fichero correspondiente distritos.xlsx que volvemos a importar :

```python
distritos=pd.read_excel('distritos.xlsx')
distritos
```

```python
distritos.distrito.value_counts()
```

```python
df_Alquiler=pd.merge(df_Alquiler,distritos[['distrito','location']],on='location',how='left')
```

```python
df_Alquiler.info()
```

```python
df_Alquiler.distrito.value_counts()
```

### Variable Temporalidad

Una vivienda alquilada de forma temporal, debería incroporar en su descripción el concepto de temporal, temporada, con un máximo de 11 meses. Buscamos las viviendas cuya descripción contenga uno de de estos conceptos :

```python
df_Alquiler['temporalidad']= np.where(df_Alquiler['description'].str.contains(r'tempora|Tempora|TEMPORA|11 meses|11 Meses|11 MESES')==True,'Temporal','Indefinido')
```

```python
df_Alquiler.temporalidad.value_counts()
```

### Variable cambio_precio

Este concepto se encuentra en la columna 'price' :

```python
df_Alquiler['cambio_precio']= np.where(df_Alquiler['price'].str.contains(r'ha bajado')==True,'Si','No')
```

```python
df_Alquiler.cambio_precio.value_counts()
```

### Variable Baños

El número de baños aparece en la columna 'features' :

```python
df_Alquiler['baños_pre']= np.where(df_Alquiler['features'].str.contains(r'habitacion')==True,
                         df_Alquiler['features'].str.split('-').str[2],
                        (np.where(df_Alquiler['features'].str.contains(r'baño')==True,
                         df_Alquiler['features'].str.split('-').str[1],'No')))
```

```python
df_Alquiler['baños_pre'].value_counts()
```

Asignamos el número de baños que corresponde, en el caso de que no haya valor, asignamos el valor 1 por defecto akl considerar que una vivienda tiene como mínimo un baño :

```python
df_Alquiler['baños']= np.where(df_Alquiler['baños_pre'].str.contains(r'baño')==True,
                          df_Alquiler['baños_pre'].str.split(' ').str[1],1)
```

```python
df_Alquiler.baños.value_counts()
```

Convertimos la variable a numérica :

```python
df_Alquiler['baños'] = df_Alquiler['baños'].astype('int')
```

```python
df_Alquiler.drop(['baños_pre'], axis=1, inplace=True)
```

```python
df_Alquiler.drop(['Unnamed: 0'], axis=1, inplace=True)
```
